# Z-test


- A Z-test is a statistical test used to determine whether there is a significant difference between the means of two groups
when the sample size is large and the population standard deviation is known. It's often used when comparing a sample mean
to a known population mean, or when comparing two independent samples.
The Z-test statistic measures how many standard deviations a data point is from the mean of the sample or population.

- The Z-test compares this Z-test statistic to a standard normal distribution (mean = 0, standard deviation = 1)
to determine the probability (p-value) of observing the sample mean if the null hypothesis (typically stating t
there is no difference between the means) is true.

- If the p-value is less than a predetermined significance level (often denoted as ), typically 0.05, 
then the null hypothesis is rejected in favor of the alternative hypothesis, suggesting that there is a 
significant difference between the means of the two groups.

-

## t-test

-  A t-test is a statistical test used to determine if there is a significant difference between the means of two groups.
   It is commonly used for comparing the means of samples from two populations when the population standard deviations 
   are unknown and the sample sizes are small (typically less than 30).

- There are several types of t-tests:

- One-sample t-test:
  This test compares the mean of a single sample to a known value (often a population mean) to determine 
  if the sample mean is significantly different from the known value.

- Independent two-sample t-test:
   This test compares the means of two independent samples to determine if they are significantly different from each other.

- Paired sample t-test: 
    Also known as a dependent sample t-test, this test compares the means of two related groups to determine 
    if there is a significant difference between them. The groups are related because they are paired or matched in some way
    (e.g., before-and-after measurements on the same subjects).
    
- The t-test calculates a t-statistic, which is a measure of the difference between the means of the two groups relative to the   variation within the groups. The larger the t-statistic, the more likely it is that the difference between the means is         statistically significant.

- The interpretation of the results typically involves comparing the calculated t-statistic to a critical value from the t-       distribution, based on the chosen significance level (often 0.05). If the calculated t-statistic exceeds the critical value,
  we reject the null hypothesis and conclude that there is a significant difference between the means of the two groups.
  
- Overall, t-tests are widely used in hypothesis testing and are particularly useful for comparing means when dealing with small sample sizes and when the population standard deviations are unknown.

# two- sample t-test (w.r.t two independent sample)

```python
from scipy.stats import ttest_ind

# Assuming 'sample1' and 'sample2' are your two independent samples
# For example:
sample1 = [1, 2, 3, 4, 5]
sample2 = [6, 7, 8, 9, 10]

t_statistic, p_value = ttest_ind(sample1, sample2)

print("P-value:", p_value)
```

```python
if p_value<=0.05:
    print('reject null hypothesis')
else:
    print('accept the null hypothesis')
```

```python
# Ex 2 
from scipy.stats import ttest_ind
import numpy as np

# Sample data for Technique A
technique_a_scores = [82, 85, 88, 90, 78, 84, 87, 89, 79, 80]

# Sample data for Technique B
technique_b_scores = [90, 92, 85, 88, 86, 91, 84, 82, 88, 87]

# Perform two-sample t-test
t_statistic, p_value = ttest_ind(technique_a_scores, technique_b_scores)

# Output results
print("P-value:", p_value)

# Interpret results
alpha = 0.05
if p_value <= alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

# paired sample t-test

```python
# Ex 1
from scipy.stats import ttest_rel
import numpy as np

# Sample data for "before" scores
before_scores = [75, 80, 85, 90, 72, 78, 83, 88, 79, 82]

# Sample data for "after" scores (corresponding to the same individuals)
after_scores = [80, 82, 88, 92, 75, 79, 85, 90, 80, 84]

# Perform paired sample t-test
t_statistic, p_value = ttest_rel(before_scores, after_scores)

# Output results
print("P-value:", p_value)

# Interpret results
alpha = 0.05
if p_value <= alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

# Chi-Square test 

- The chi-square test is a statistical test used to determine whether there is a significant association between categorical variables. It is commonly used when you have categorical data and want to assess whether the observed frequencies of different categories differ from what would be expected under the null hypothesis of no association.

- There are different types of chi-square tests depending on the specific scenario:

- Chi-square test for independence: This test is used to determine whether there is a significant association between two categorical variables. It compares the observed frequencies of the categories with the expected frequencies under the assumption that the two variables are independent.

- Chi-square test for goodness of fit: This test is used to determine whether the observed frequencies of a single categorical variable differ significantly from the expected frequencies specified by a theoretical distribution or hypothesis.
- 

- 

```python
# Ex 1
from scipy.stats import chi2_contingency

# Create a contingency table (observed frequencies)
observed = [[10, 20, 30],
            [15, 25, 35],
            [25, 30, 40]]

# Perform chi-square test for independence
chi2_statistic, p_value, _, _ = chi2_contingency(observed)

# Output results
print("Chi-square statistic:", chi2_statistic)
print("P-value:", p_value)

# Interpret results
alpha = 0.05
if p_value <= alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```
